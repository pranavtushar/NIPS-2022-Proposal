```python
# Data Manipulation
import numpy as np
import pandas as pd

# Data
from data import RealtimeFinanceData

# ML
from time2vec import Time2Vec
```

### Amazon stock price prediction

```python
data_instance = RealtimeFinanceData(name="AMZN")
data_instance.plot_close_price()
X_train, X_test, y_train, y_test = data_instance.generate_data(SEQ_LEN=50)
```

```python
model_instance = Time2Vec(dim=50)
model_instance.set_LSTM()
```

```python
model_instance.train(X_train, y_train)
```

```python
model_instance.plot_training()
```

```python
model_instance.evaluate(X_test, y_test)
```

```python
y_hat = model_instance.model.predict(X_test)
```

```python
model_instance.plot_predictions(y_test, y_hat, item="AMAZON stock price")
```
